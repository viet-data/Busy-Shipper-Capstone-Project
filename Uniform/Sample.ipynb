The library data use to load all information for Uniform cost search algorithm

data.get_data(get_data(filename, address, dist, graph=True)

Paramerters: 

             filename: string

             address: List,
                 List of Latitude and Longitude 
             
             dist: int, 
                 The radius
             
             graph: bool, optional
                 If True, print the graph of map
                 
Returns:Graph G, start position s, maximum number of packages m, distance matrix mat_dist, list of orders, list of customer, sto and initial positons lst, dictionary to find index on mat_dist idx, list of customers, list of stores sto
    
            

```python
import data
import pickle

with open('AMS.pickle', 'rb') as f:
    G = pickle.load(f)
    
filename = 'AMS_4_8.txt'
s, m, mat_dist, lst, idx, orders, cus, sto = data.get_data(filename, G)
```

```python
%load_ext autoreload
%autoreload 2
import problem
import Uniform

P = problem.problem(cus, sto, initial_position=s, orders=orders[:], G=G, mat_dist=mat_dist, idx=idx, max_packages=m)
path, cost = Uniform.uniform_cost_search(P)
cost
```
