```python
%load_ext autoreload
%autoreload 2
import BGS
import data
import pickle

with open('Catalunya.pickle', 'rb') as f:
    G = pickle.load(f)
    
filename = 'CAT_200_2.txt'
s, m, mat_dist, lst, idx, orders, cus, sto = data.get_data(filename, G)
```

The function BGS.BGS is called to solve the problem.

Parameters:

            cus, list
            sto, list
            s, int
            m, int
            G, Graph
            mat_dist, array
            limit_num, int
            limit_time, int
Returns:
    
        cost: Minimum cost
        paths: A list of path that have minimum cost
        results: A list of solutions in many tries 
            

```python
%%time
cost, paths, results = BGS.BGS(cus, sto, s, m, G, mat_dist, idx, limit_num=60, limit_time=60, num_gen=20, num_pop=300)
print('Final minimum cost:', cost)
```
